# Exploratory Analysis of Financial Transactions

**Domain:** Data Science  
**Project:** Minor Project  
**Dataset:** Credit Card Transactions Dataset  
**Objective:** Explore financial transaction data to identify trends, distributions, and suspicious fraud patterns.

## 1. Import Required Libraries
We will use Pandas for data analysis and Matplotlib for visualization.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (8, 5)
```

## 2. Load Dataset
Place the downloaded Kaggle file at `data/creditcard.csv` before running this notebook.

```python
DATA_PATH = "data/creditcard.csv"

df = pd.read_csv(DATA_PATH)
df.head()
```

## 3. Basic Dataset Understanding
Here we check the size, columns, data types, and statistical summary of the dataset.

```python
print("Dataset Shape:", df.shape)
print("Columns:")
print(df.columns.tolist())
```

```python
df.info()
```

```python
df.describe()
```

## 4. Missing Values and Duplicate Records
Before analysis, we check whether the dataset has missing or duplicate values.

```python
missing_values = df.isnull().sum()
print("Total Missing Values:", missing_values.sum())
missing_values[missing_values > 0]
```

```python
duplicate_count = df.duplicated().sum()
print("Duplicate Rows:", duplicate_count)
```

## 5. Understanding the Target Column
The `Class` column tells whether a transaction is normal or fraudulent.

- `0` = Normal transaction
- `1` = Fraud transaction

```python
class_counts = df["Class"].value_counts()
class_percent = df["Class"].value_counts(normalize=True) * 100

print("Class Counts:")
print(class_counts)
print("
Class Percentage:")
print(class_percent)
```

```python
class_counts.plot(kind="bar")
plt.title("Fraud vs Normal Transactions")
plt.xlabel("Class: 0 = Normal, 1 = Fraud")
plt.ylabel("Number of Transactions")
plt.xticks(rotation=0)
plt.show()
```

### Observation
The dataset is highly imbalanced because normal transactions are much higher than fraudulent transactions.

## 6. Transaction Amount Distribution
This chart shows how transaction amounts are distributed.

```python
plt.hist(df["Amount"], bins=50)
plt.title("Transaction Amount Distribution")
plt.xlabel("Transaction Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
print("Amount Summary:")
print(df["Amount"].describe())
```

## 7. Transaction Time Distribution
The `Time` column represents the seconds elapsed between transactions.

```python
plt.hist(df["Time"], bins=50)
plt.title("Transaction Time Distribution")
plt.xlabel("Time in Seconds")
plt.ylabel("Number of Transactions")
plt.show()
```

## 8. Fraud vs Normal Transaction Amount Comparison
Now we separate normal and fraud transactions to compare their amount behavior.

```python
normal = df[df["Class"] == 0]
fraud = df[df["Class"] == 1]

print("Normal Transactions:", normal.shape[0])
print("Fraud Transactions:", fraud.shape[0])
```

```python
print("Normal Transaction Amount Summary:")
print(normal["Amount"].describe())

print("
Fraud Transaction Amount Summary:")
print(fraud["Amount"].describe())
```

```python
plt.boxplot([normal["Amount"], fraud["Amount"]], labels=["Normal", "Fraud"], showfliers=False)
plt.title("Normal vs Fraud Transaction Amount")
plt.ylabel("Transaction Amount")
plt.show()
```

### Observation
The boxplot helps compare the spread of normal and fraudulent transaction amounts. `showfliers=False` is used because extreme outliers can make the plot difficult to read.

## 9. Suspicious Pattern Exploration
Fraud transactions are rare, so we inspect their amount and time behavior separately.

```python
plt.scatter(fraud["Time"], fraud["Amount"], alpha=0.5)
plt.title("Fraud Transactions Over Time")
plt.xlabel("Time")
plt.ylabel("Amount")
plt.show()
```

```python
plt.hist(fraud["Amount"], bins=30)
plt.title("Fraud Transaction Amount Distribution")
plt.xlabel("Fraud Transaction Amount")
plt.ylabel("Frequency")
plt.show()
```

## 10. Correlation with Fraud Class
Correlation helps identify which numerical features have stronger relationships with the `Class` variable.

```python
correlation_with_class = df.corr(numeric_only=True)["Class"].sort_values()
correlation_with_class
```

```python
top_corr = correlation_with_class.drop("Class").abs().sort_values(ascending=False).head(10)

top_corr.plot(kind="bar")
plt.title("Top 10 Features Correlated with Fraud Class")
plt.xlabel("Feature")
plt.ylabel("Absolute Correlation")
plt.xticks(rotation=45)
plt.show()
```

## 11. Final Project Summary
Fill these after running the notebook.

```python
total_transactions = len(df)
fraud_transactions = len(fraud)
normal_transactions = len(normal)
fraud_percentage = (fraud_transactions / total_transactions) * 100

print("Final Summary")
print("-------------")
print("Total Transactions:", total_transactions)
print("Normal Transactions:", normal_transactions)
print("Fraud Transactions:", fraud_transactions)
print("Fraud Percentage:", round(fraud_percentage, 4), "%")
print("Average Transaction Amount:", round(df["Amount"].mean(), 2))
print("Maximum Transaction Amount:", round(df["Amount"].max(), 2))
```
